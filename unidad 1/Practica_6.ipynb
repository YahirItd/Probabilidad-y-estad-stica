```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
from matplotlib import pyplot
```

```python
Auto= pd.read_csv("autos2.csv",encoding="ISO-8859-1")
```

```python
Auto.head()
```

```python
Auto.info()
```

Determinar el valor mínimo de precios

Determinar el valor máximo de precios

Determinar el valor de la moda de precios

Determinar el valor de la media de precios

Determinar el valor de la mediana de precios

Determinar el valor de la desviación estándar de precios

Determinar los estadísticos y cuartiles con la función describe

```python
maximo=Auto.price.max()
minimo=Auto.price.min()
moda=Auto.price.mode()
media=Auto.price.mean()
mediana=Auto.price.median()
desviacion_estandar= Auto.price.std()
estadistica=Auto.price.describe()
```

```python
print("Valor mínimo:", minimo,"\n Valor máximo:",maximo,"\n Moda:",moda,
      "\n Mediana:",mediana,"\n Desviación estándar:",desviacion_estandar,"\n Estadistica con funcion describe:",estadistica)
```

```python
import matplotlib.pyplot as plt
import math
k= int(1+ math.log2(len(Auto.price)))
plt.hist(Auto.price, alpha=0.5, bins=k)
plt.title("Histograma de los precios")
plt.xlabel("Precios")
plt.ylabel("Frecuencia")
```


Se puede apreciar cómo el estándar de los precios suele estar en el rango de 5000, con una mayor concentración en ese punto, mientras que a medida que los precios aumentan, la frecuencia disminuye notablemente. Se observa una menor cantidad de valores en el rango de 20000 en adelante, llegando a ser muy escasos alrededor de 45000.

```python
ax1=sns.boxplot(x="body-style",y="price",data=Auto)
ax1.set_title('Relacion con el precio y tipo de vehículo')
```

En esta gráfica se puede apreciar que el tipo de vehículo que parece ser más económico es el hatchback, con una menor dispersión de precios y valores más bajos en general. En un rango medio de precios se encuentra el sedán, con algunos valores atípicos que superan el promedio. Por otro lado, los vehículos más costosos corresponden al tipo hardtop, que presenta una mayor variabilidad y precios máximos más elevados en comparación con el resto.

```python
ax2=sns.boxplot(x="drive-wheels",y="price",data=Auto)
ax2.set_title('Relacion con el precio y tipo de ruedas')
```

En esta gráfica se aprecia que el tipo de tracción más económico es el fwd, con precios más bajos y menor dispersión. Le sigue el 4wd, con un rango de precios ligeramente superior pero aún dentro de un nivel moderado. Finalmente, el rwd es el que presenta los precios más elevados, con una mayor variabilidad y valores atípicos que alcanzan los niveles más altos.

```python
ax3=sns.boxplot(x="engine-location",y="price",data=Auto)
ax3.set_title('Relacion con el precio y ubicación del motor')
```

En esta grafica se muestra mas variabilidad de costos cuando el motor esta adelante y es notablemente mas caro y con menos variabilidad cuando se ubica por atras el motor
