```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
from scipy.stats import pearsonr
```

**Grupo 1**

```python
g1=pd.read_csv("/grupo1.csv")
g1.head()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(g1['Calificacion'], kde=True, bins=10, color='skyblue')
plt.title('Histograma de Calificaciones Finales')
plt.xlabel('Calificacion')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
X = g1[['HorasEstudio']]
y = g1['Calificacion']

# Crear el modelo
modelo = LinearRegression()
modelo.fit(X, y)

# Obtener pendiente e intercepto
pendiente = modelo.coef_[0]
intercepto = modelo.intercept_

print(f'Ecuación de regresión: y = {pendiente:.2f}x + {intercepto:.2f}')
```

```python
r, _ = pearsonr(g1['HorasEstudio'], g1['Calificacion'])
print(f'Coeficiente de correlación de Pearson: r = {r:.3f}')
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='HorasEstudio', y='Calificacion', data=g1, color='blue', label='Datos')

x_vals = np.array(plt.gca().get_xlim())
y_vals = intercepto + pendiente * x_vals
plt.plot(x_vals, y_vals, '--', color='red', label='Línea de regresión')

plt.title('Relación entre Horas de Estudio y Calificación Final')
plt.xlabel('Horas de Estudio')
plt.ylabel('Calificación Final')
plt.legend()
plt.grid(True)
plt.show()
```

1. **¿Cuál es la interpretación del coeficiente de regresión?**

La pendiente de la regresión es aproximadamente -0.11, lo que significa que:

Por cada hora adicional de estudio, la calificación final disminuye en promedio 0.11 puntos.

Esto es contrario a lo esperado (uno pensaría que más estudio = mejor calificación), por lo tanto sugiere que no hay una relación positiva entre el tiempo de estudio y el rendimiento académico en este grupo.

---

2. **¿Cómo se interpreta el coeficiente de correlación de Pearson?**

El coeficiente de correlación r = -0.057 indica que:

La relación es muy débil (cercana a 0)

En resumen: no existe una correlación lineal significativa entre las horas de estudio y las calificaciones finales en este grupo.

---

3. **¿El modelo de regresión lineal es estadísticamente significativo?**
 El modelo no es estadísticamente significativo.

 ---

4. **¿Qué tan confiable es el modelo de regresión para predecir las calificaciones?**

El modelo es poco confiable para predecir calificaciones a partir del tiempo de estudio en este grupo. No captura una relación clara ni útil.
