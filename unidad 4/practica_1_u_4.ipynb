```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
media= 70
desv_std=4
num_datos=500
datos=np.random.normal(media,desv_std,num_datos)
datos[0:50]
```

```python
plt.figure(figsize=(6,4))
plt.hist(datos,bins=15,alpha=0.4,color='b')
plt.title('Distribución de datos')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

```python
x= np.linspace(media - 3*desv_std,media + 3*desv_std,num_datos)
pdf= 1/ (desv_std * np.sqrt(2 * np.pi)) * np.exp( - (x - media)**2 / (2 * desv_std**2))
```

```python
pdf.size,num_datos
```

```python
print("Imprimir los primeros 50 valores de x",x[0:50]," , y los primeros 50 pdf",pdf[0:50])
```

```python
plt.figure(figsize=(6,4))
plt.plot(x,pdf,label='Normal:(media:{};ds:{})'.format(np.round(media,2),np.round(desv_std,2)))
plt.title ('Función de Densidad de Probabilidad (PDF) - Distibución Normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.ylabel('Densidad de probabilidad')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.hist(datos,bins=15,alpha=0.4, density=True ,color='b')
plt.plot(x,pdf,'k',linewidth=2)
plt.title('Distribución normal')
plt.xlabel('Peso en kg de los estudiantes')
plt.grid(True)
plt.show()
```

# Ejercicio 2

Una revision de los pesos de 500 estudiantes del ITD, evidenció que la media es de 70 kilogramos y la desviacion estandar es de 4 kilogramos.

Suponiendo que los valores se distribuyen normalmente. Hallar lo siguiente:

Cuantos estudiantes pesan menos de 75 kg.

Cuantos estudiantes pesan menos de 62 kg.

Cuantos estudiantes pesan mas de 79 kg.

Primero necesitamos calcular la probabilidad de que un estudiante pese menos de 75 kg utilizando la función de distribución
acumulativa (CDF) de la distribución normal

```python
plt.figure(figsize=(6,4))
# Parámetros de la distribución norma L
media= 70
desv_std = 4
total_estudiantes = 500
peso_limite =79
prob_menos_75kg =norm.cdf(peso_limite, media, desv_std)
```

```python
estudiantes_menos_75kg = total_estudiantes * prob_menos_75kg
print("Probabilidad de que un estudiante pese menos de 75 kg:", prob_menos_75kg)
print("Número de estudiantes que pesan menos de 75 kg:", estudiantes_menos_75kg)
#Graficar la distribución normal
x = np.linspace(media - 4 * desv_std, media +4 * desv_std, 1000)
y= norm.pdf(x, media, desv_std)
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between (x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 75 kg. \n $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
#Parámetros de la distribución normal
media = 70 # Media
desv_std = 4 # Desviación estándar
total_estudiantes = 500
# VARIABLE para Cálculo de La probabilidad de que un estudiante pese menos de 62 kg
peso_limite = 62
# Primero necesitamos calcular La probabilidad de que un estudiante pese menos
# de 75 kg utilizando La función de distribución acumulativa (CDF) de la distribución normal
prob_menos_62kg = norm.cdf(peso_limite, media, desv_std)
# Cálculo del número de estudiantes que pesan menos de 75 kg
estudiantes_menos_62kg = total_estudiantes * prob_menos_62kg
print("Probabilidad de que un estudiante pese menos de 62 kg:", prob_menos_62kg)
print("Número de estudiantes que pesan menos de 62 kg:", estudiantes_menos_62kg)

# Calcular la distribución normal
x = np.linspace(media -4* desv_std, media + 4* desv_std, total_estudiantes)
y= norm.pdf(x, media, desv_std)
# Graficar la distribución normal
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between (x, y, where=(x <= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese menos de 62 kg, \n $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6,4))
# Parámetros de La distribución normal
media = 70 # Media
desv_std = 4 # Desviación estándar
peso_limite = 79 # Peso Limite
total_estudiantes = 500
# Calculamos La probabilidad de que un estudiante pese menos o igual a 79 kg
prob_menos_79kg = norm.cdf(peso_limite, media, desv_std)
# La probabilidad de que un estudiante pese más de 79 kg es complementaria
prob_mas_79kg = 1 -prob_menos_79kg
# Calculamos el número aproximado de estudiantes que pesan más de 79 kg
estudiantes_mas_79kg = total_estudiantes * prob_mas_79kg
print("Probabilidad de que un estudiante pese más de 79 kg:", prob_mas_79kg)
print("Número aproximado de estudiantes que pesan más de 79 kg:", estudiantes_mas_79kg)

# Calcular La función de densidad de probabilidad de la distribución normal
x = np.linspace(media -4* desv_std, media + 4* desv_std, total_estudiantes)
y= norm.pdf(x, media, desv_std)
plt.plot(x, y, 'b', linewidth=2)
plt.fill_between(x, y, where=(x >= peso_limite), color='skyblue', alpha=0.5)
plt.title('Probabilidad de que un estudiante pese mas de 79 Kg. \n $\mu=70$, $\sigma=4$, X=500')
plt.xlabel('Peso (kg)')
plt.ylabel('Densidad de probabilidad')
plt.grid(True)
plt.show()
```
